```python
def convert_string_to_date_time(date_str):
    date_ = datetime.strptime(date_str, '%Y-%m-%d').date()

    
    return date_


def get_year(x):
    return x.year

def get_month(x):
    return x.month

def get_day(x):
    return x.day


```
